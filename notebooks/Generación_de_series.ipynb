```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Funcin para generar una serie con VARIANZA CONSTANTE
def serieVarianzaConstante(num_segmentos, tamano_segmento):
    serie = np.concatenate([
        np.random.normal(np.random.randint(-10, 10), 1, tamano_segmento)  
        for _ in range(num_segmentos)  
    ])
    return pd.Series(serie)

# Funcin para generar una serie con VARIANZA VARIABLE
def serieVarianzaVariable(num_segmentos, tamano_segmento):
    serie = np.concatenate([
        np.random.normal(0, np.random.randint(1, 50), tamano_segmento)  
        for _ in range(num_segmentos)  
    ])
    return pd.Series(serie)  

# Generar las dos series usando las funciones
serie1 = serieVarianzaConstante(10, 1000)
serie2 = serieVarianzaVariable(10, 1000)


# Graficar la serie con varianza constante
plt.figure(figsize=(16, 4)) 
plt.plot(serie1) 
plt.title("Serie temporal con varianza constante") 
plt.show() 

# Graficar la serie con varianza variable
plt.figure(figsize=(16, 4))
plt.plot(serie2)  
plt.title("Serie temporal con varianza variable") 
plt.show()    
```
