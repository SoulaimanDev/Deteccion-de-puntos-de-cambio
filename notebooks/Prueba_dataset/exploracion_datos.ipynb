##  Visualizaci贸n de las se帽ales para ts_id = 0 (indoor)

```python
#  Definir las columnas de metadatos
columnas_metadatos = [
    'ts_id', 'ts_challenge_id', 'group', 'subject',
    'routine', 'change_points', 'activities', 'split'
]

# Seleccionar un ejemplo (primera fila)
ejemplo = df_flat.iloc[0]

# Mostrar los metadatos de este ejemplo
metadatos_ejemplo = ejemplo[columnas_metadatos]
print("\n Metadatos del ejemplo seleccionado:\n")
print(metadatos_ejemplo)
```

```python
# Definir las columnas de se帽ales (a partir de la columna 8)
columnas_senales = df_flat.columns[8:]

# Seleccionar un ejemplo (por ejemplo, la fila 0)
ejemplo = df_flat.iloc[0]

# Iterar sobre cada columna de se帽al
for columna in columnas_senales:
    senal = ejemplo[columna]

    # Verificar si es una cadena y limpiar los corchetes
    if isinstance(senal, str):
        senal = senal.strip("[]")
        if senal == "":
            continue

        # Convertir la cadena en una lista de valores float
        valores = [float(x) for x in senal.split(",") if x.strip() != ""]

        # Generar los valores de tiempo en segundos (frecuencia de muestreo: 50 Hz  0.02 s)
        valores_tiempo = [i * 0.02 for i in range(len(valores))]

        # Crear la figura
        plt.figure(figsize=(15, 4))
        plt.plot(valores_tiempo, valores)
        plt.title(f"TS#1 - Actividades: {ejemplo['activities']}")
        plt.xlabel("Tiempo (s)")
        plt.ylabel(columna)
        plt.grid(False)
        plt.show()

```
